```python
import pandas as pd
import numpy as np
import os
```

```python
file_info = "Resources/purchase_data.csv"
purchase_df = pd.read_csv(file_info)
purchase_df.head()
```

```python
# given data facts not in csv

active_players = 1163

active_malepct = .84
active_femalepct = .14
active_NApct = .02

active_2024 = .448
active_1519 = .186
active_2529 = .134
active_rest = .232
```

```python
purchases_players = purchase_df['SN'].nunique()
print(purchases_players)
```

```python
players_df = pd.DataFrame([{'Total Players': purchases_players}])
players_df
```

```python
purchases_total = len(purchase_df['SN'])
print(purchases_total)
```

```python
purchases_items = purchase_df['Item Name'].nunique()
print(purchases_items)
```

```python
purchases_avgprice = round(purchase_df['Price'].mean(), 2)
print(f'${purchases_avgprice}')
```

```python
purchases_revenue = round(purchase_df['Price'].sum(), 2)
print(f'${purchases_revenue}')
```

```python
summary_items_df = pd.DataFrame([{'# of unique items sold': purchases_items,
                    'average price': f"${purchases_avgprice}",
                    '# of total purchases': purchases_total,
                    'total revenue': f"${purchases_revenue}"}])
summary_items_df
```

```python
gender_df = purchase_df.loc[:, ['SN', "Gender",]]
gender_df2 = gender_df.drop_duplicates(['SN'])
gender_df2 = gender_df2.groupby(['Gender'], as_index=False).agg({'SN':'count'})
gender_df2 = gender_df2.sort_values("SN", ascending=False).reset_index(drop=True)

gender_df2['% of players'] = (gender_df2['SN'] / purchases_players * 100).map("% {:.2f}".format)
gender_df2 = gender_df2.rename(columns={'SN':'Count'})

gender_df2
```

```python
g_purch_df = purchase_df.copy()
g_purch_df = g_purch_df.drop(["Item Name" ,'Item ID' ,"Age" ,"Purchase ID"], axis=1)
g_purch_df = g_purch_df.groupby(['Gender'], as_index=False).agg({'SN' : 'count', 'Price' : 'sum'})
g_purch_df = g_purch_df.rename(columns={'SN':'Purchase Count'})
g_purch_df = g_purch_df.rename(columns={'Price':'Total Purchase Value'})
g_purch_df = g_purch_df.sort_values('Purchase Count', ascending=False).reset_index(drop=True)

g_purch_df['Average Purchase Price'] = round((g_purch_df['Total Purchase Value'] / g_purch_df['Purchase Count']), 2)
g_purch_df['Avg Total Purchase per Person'] = round((g_purch_df['Total Purchase Value'] / gender_df2['Count']), 2)

g_purch_df
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, purchase_df['Age'].max()]
bin_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_bracket_df = purchase_df.copy()
age_bracket_df2 = age_bracket_df.drop(["Item Name" ,'Item ID' ,"Gender" ,"Purchase ID", 'Price'], axis=1)
age_bracket_df2 = age_bracket_df2.drop_duplicates(["SN"])
age_bracket_df2['Age Bracket'] = pd.cut(age_bracket_df2['Age'], bins, labels=binNames, include_lowest=True)
age_bracket_df2 = age_bracket_df2.groupby(['Age Bracket']).agg({'SN' : 'count'})
age_bracket_df2 = age_bracket_df2.rename(columns={'SN':'Count'})
age_bracket_df2['% of players'] = (age_bracket_df2['Count'] / purchases_players * 100).map("{:.2f} %".format)

age_bracket_df2
```

```python
a_purch_df = age_bracket_df.copy()
a_purch_df = a_purch_df.drop(["Item Name" ,'Item ID' ,"Gender" ,"Purchase ID"], axis=1)
a_purch_df['Age Bracket'] = pd.cut(a_purch_df['Age'], bins, labels=binNames, include_lowest=True)
a_purch_df = a_purch_df.sort_values('Age Bracket', ascending=True).reset_index(drop=True)
a_purch_df = a_purch_df.groupby(['Age Bracket'], as_index=False).agg({'SN' : 'count', 'Price' : 'sum'})

a_purch_df = a_purch_df.rename(columns={'SN':'Purchase Count'})
a_purch_df = a_purch_df.rename(columns={'Price' : 'Total Purchase Value'})

a_purch_df['Average Purchase Price'] = round((a_purch_df['Total Purchase Value'] / a_purch_df['Purchase Count']), 2)
a_purch_df['Avg Total Purchase per Person'] = round((a_purch_df['Total Purchase Value'] / age_bracket_df2['Count']), 2)

a_purch_df
```

```python
top_spenders_df = purchase_df.copy()
top_spenders_df2 = top_spenders_df
top_spenders_df2["Avg"] = top_spenders_df2['Price']
top_spenders_df2 = top_spenders_df2.groupby(["SN"], as_index=False).agg({'Purchase ID' : 'count', 'Price' : 'sum', 'Avg' : 'mean'})
top_spenders_df3 = top_spenders_df2.sort_values("Price", ascending=False)
top_spenders_df3 = top_spenders_df3.rename(columns={'Price':'Total Purchase Value'})
top_spenders_df3 = top_spenders_df3.rename(columns={'Avg':'Average Purchase Price'})
top_spenders_df3['Average Purchase Price'] = round(top_spenders_df3['Average Purchase Price'], 2)
top_spenders_df3 = top_spenders_df3.rename(columns={'Purchase ID':'Purchase Count'})

top_spenders_df3.head()
```

```python
item_df = purchase_df
item_df2 = item_df
item_df2 = item_df2.drop(['Purchase ID', 'SN', 'Age', 'Gender'], axis=1).copy()
item_df2 = item_df2.groupby(['Item Name'], as_index=False).agg({'Item ID':'count', 'Price':'mean'})
item_df2['Total Purchase Value'] = round((item_df2['Item ID'] * item_df2['Price']), 2)
item_df2['Price'] = round(item_df2['Price'], 2)
item_df2 = item_df2.sort_values("Total Purchase Value", ascending=False)
item_df2 = item_df2.rename(columns={'Item ID':"Purchase Count"})
item_df2 = item_df2.rename(columns={'Price':'Item Price'})

item_df2
```
